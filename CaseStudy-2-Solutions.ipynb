```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1=pd.read_csv("C:\Dump\Ishwar\Data_Science\Edureka\Class 4\Dataset\MathScoreTerm1.csv")
df2=pd.read_csv("C:\Dump\Ishwar\Data_Science\Edureka\Class 4\Dataset\DSScoreTerm1.csv")
df3=pd.read_csv("C:\Dump\Ishwar\Data_Science\Edureka\Class 4\Dataset\PhysicsScoreTerm1.csv")
```

```python
df1.drop("Name",axis=1,inplace=True)
df2.drop("Name",axis=1,inplace=True)
df3.drop("Name",axis=1,inplace=True)
```

```python
dic1={
    'White American': 1,
    'European American': 2,
    'Hispanic': 3,
    'African American': 4
}

dic2={
    'M': 1,
    'F': 2
}
```

```python
df1.isnull().sum()
df2.isnull().sum()
df3.isnull().sum()
```

```python
df1["Ethinicity"].replace(dic1,inplace=True)
df2["Ethinicity"].replace(dic1,inplace=True)
df3["Ethinicity"].replace(dic1,inplace=True)
```

```python
df1["Sex"].replace(dic2,inplace=True)
df2["Sex"].replace(dic2,inplace=True)
df3["Sex"].replace(dic2,inplace=True)
```

```python
df1["Score"].fillna(df1["Score"].mean(),inplace=True)
df2["Score"].fillna(df2["Score"].mean(),inplace=True)
df3["Score"].fillna(df3["Score"].mean(),inplace=True)
```

```python
df = pd.concat([df1,df2,df3],axis=1)
```

```python
df.T.drop_duplicates().T
```

```python
newdf=df.T.drop_duplicates().T
```
